```python
from keras.models import load_model
from keras.preprocessing.image import load_img
import numpy as np
from keras.preprocessing import image
import cv2
img_dims = 150
```

```python
model=load_model('Models/Chest_X-Ray-A.h5')
```

```python
img = image.load_img('../chest_xray/chest_xray/train/PNEUMONIA/person255_bacteria_1188.jpeg', target_size=(150,150))
x = image.img_to_array(img)
x = np.expand_dims(x, axis =0)
x = x.astype('float32') / 255
```

```python
 images = np.vstack([x])
```

```python
images
```

```python
classes = model.predict(images, batch_size = 10)
```

```python
classes
```

```python
print(classes[0])
```

```python
import matplotlib.pyplot as plt

if classes[0]> 0.5:
    print( 'cell  is pneumonia')
    plt.imshow(img)
else:
    print("cell " + 'is normal')
plt.imshow(img)
```
